```python
fig, axs= plt.subplots(nrows=2, ncols=2, figsize=(6,6), dpi=100, facecolor='white', constrained_layout=True)

### set font
font = {'family' : 'Arial',
        'weight' : 'normal',
        'size'   : 8}
mpl.rc('font', **font)  # pass in the font dict as kwargs

### A ###
# set data
x = np.arange(0, 40, 1)
y7 = 1.152**x
y8 = 1.204**x
y9 = 1.282**x
axs[0,0].set_title('a', fontweight='bold', fontsize=10, loc='left', pad=10, x=-0.05)
# combined effect
axs[0,0].plot(x, y9, label='high', color='#2ca02c')
axs[0,0].plot(x, y8, label='medium', color='#ff7f0e')
axs[0,0].plot(x, y7, label='low', color='#1f77b4')
axs[0,0].spines['right'].set_visible(False)
axs[0,0].spines['top'].set_visible(False)
axs[0,0].tick_params(axis='both', length=2, pad=1)
axs[0,0].set_xlabel('Slope (S) ()')
axs[0,0].set_ylabel('Physical Erosion (Ep)')
axs[0,0].annotate('Ep ' + r'$\varpropto$' + ' (PI$^S$)', (25,25))
axs[0,0].set_ylim(bottom=0, top=200)
axs[0,0].set_xlim(left=0, right=40)
axs[0,0].legend(title='Precipitation Intensity (PI):')._legend_box.align='left'
# axs[0,0].set_xticks([])
axs[0,0].set_yticks([])

### B ###
# prepare data
x = np.arange(0,40,0.1)
y4 = 50 - 1.1**x
y5 = 30 - 1.1**x
y6 = 10 - 1.1**x
y4_1 = 50 - 1.13**x
y5_1 = 30 - 1.13**x
y6_1 = 10 - 1.13**x
# plotting
axs[0,1].set_title('b', fontweight='bold', fontsize=10, loc='left', pad=10, x=-0.05)
axs[0,1].plot(x, y4, label='more', color='#2ca02c')
axs[0,1].plot(x, y5, label='medium', color='#ff7f0e')
axs[0,1].plot(x, y6, label='less', color='#1f77b4')
axs[0,1].plot(x, y4_1, label='(PI: high)', color='#2ca02c', linestyle='--', linewidth=1)
axs[0,1].plot(x, y5_1, label='(PI: high)', color='#ff7f0e', linestyle='--', linewidth=1)
axs[0,1].plot(x, y6_1, label='(PI: high)', color='#1f77b4', linestyle='--', linewidth=1)
axs[0,1].set_ylim(bottom=0, top=100)
axs[0,1].set_xlim(left=-0, right=40)
axs[0,1].legend(title='Precipitation (P):', ncol=2, loc='upper left')._legend_box.align='left'
axs[0,1].set_xlabel('Slope (S) ()')
axs[0,1].set_ylabel('Vegetation Cover (V)')
axs[0,1].annotate('V ' + r'$\varpropto$' + ' (P) - (Ep)', (25,40))
axs[0,1].spines['right'].set_visible(False)
axs[0,1].spines['top'].set_visible(False)
axs[0,1].tick_params(axis='both', length=2, pad=1)
# axs[0,1].set_xticks([])
axs[0,1].set_yticks([])

### C ###
x = np.arange(0,40,0.1)
y4 = 50 - 1.12**x
y5 = 30 - 1.12**x
y6 = 10 - 1.12**x
y4_1 = 50 - 1.15**x
y5_1 = 30 - 1.15**x
y6_1 = 10 - 1.15**x
axs[1,0].set_title('c', fontweight='bold', fontsize=10, loc='left', pad=10, x=-0.05)
axs[1,0].plot(x, y4, label='more', color='#2ca02c',)
axs[1,0].plot(x, y5, label='medium', color='#ff7f0e')
axs[1,0].plot(x, y6, label='less', color='#1f77b4')
axs[1,0].plot(x, y4_1, label='(PI: high)', color='#2ca02c', linestyle='--', linewidth=1)
axs[1,0].plot(x, y5_1, label='(PI: high)', color='#ff7f0e', linestyle='--', linewidth=1)
axs[1,0].plot(x, y6_1, label='(PI: high)', color='#1f77b4', linestyle='--', linewidth=1)
axs[1,0].set_ylim(bottom=0, top=100)
axs[1,0].set_xlim(left=-0, right=40)
axs[1,0].tick_params(axis='both', length=2, pad=1)
axs[1,0].legend(title='Vegetation Cover (V):', ncol=2, loc='upper left')._legend_box.align='left'
axs[1,0].set_xlabel('Slope (S) ()')
axs[1,0].set_ylabel('Chemical Erosion (Ec)')
axs[1,0].annotate('Ec ' + r'$\varpropto$' + ' (V) - (Ep)', (25,40))
axs[1,0].spines['right'].set_visible(False)
axs[1,0].spines['top'].set_visible(False)
# axs[1,0].set_xticks([])
axs[1,0].set_yticks([])

### D ###
x = np.arange(0, 40, 0.1)
y7 = 1.152**x
y8 = 50 - 1.12**x
x1 = np.arange(0, math.log(50,1.12), 0.1)
y9 = 1.152**x1 + 50 - 1.12**x1
x2 = np.arange(math.log(50,1.12), 40, 0.1)
y10 = 1.152**x2
axs[1,1].set_title('d', fontweight='bold', fontsize=10, loc='left', pad=10, x=-0.05)
# combined effect
axs[1,1].plot(x, y7, label='Ep')
axs[1,1].plot(x, y8, label='Ec', color='#ff7f0e')
axs[1,1].plot(x1, y9, color='black', linestyle=':', linewidth=2, label='E')
axs[1,1].plot(x2, y10,  color='black', linestyle=':', linewidth=2)
axs[1,1].vlines(x=24.9,ymin=-2, ymax=1.152**25, colors='red', linestyles=':', linewidth=1)
axs[1,1].legend(loc='upper left')._legend_box.align='left'
axs[1,1].set_ylim(bottom=0, top=200)
axs[1,1].set_xlim(left=0, right=40)
axs[1,1].tick_params(axis='both', length=2, pad=1)
axs[1,1].spines['right'].set_visible(False)
axs[1,1].spines['top'].set_visible(False)
axs[1,1].set_xlabel('Slope (S) ()')
axs[1,1].set_ylabel('Erosion (E)')
axs[1,1].annotate('E = Ep + Ec', (2,80))
# axs[1,1].set_xticks([])
axs[1,1].set_yticks([])

# plt.savefig("Results/240131/figure3.svg", dpi=300)
```
